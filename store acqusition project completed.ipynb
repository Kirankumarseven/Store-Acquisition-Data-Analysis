```python
transactions = pd.read_csv('project_transactions.csv', 
                           dtype={'DAY':'int16','QUANTITY':'int32','WEEK_NO':'int8','household_key':'int16','SALES_VALUE':'float32'},
                          usecols= ['DAY','QUANTITY','WEEK_NO','household_key','PRODUCT_ID','SALES_VALUE'])
transactions.info(memory_usage='deep')
```

```python
transactions
```

```python
transactions.assign(
    date = pd.to_datetime('2016', format = '%Y')+ pd.to_timedelta(transactions['DAY'].sub(1).astype(str) +" days"))
```

```python
transactions = transactions.assign(
                   date = pd.to_datetime('2016', format = '%Y')+ pd.to_timedelta(transactions['DAY'].sub(1),'D')).drop('DAY', axis = 1)
```

```python
transactions.date.dtype
```

```python
transactions.date.dt.month
```

```python
transactions.head()
```

```python
(transactions.pivot_table(index = transactions['date'].dt.month, 
                          columns = transactions['date'].dt.year, 
                          values= 'SALES_VALUE', 
                          aggfunc='sum').plot())
```

```python
(transactions.pivot_table(index = transactions['date'].dt.month, 
                          columns = transactions['date'].dt.year, 
                          values= 'SALES_VALUE', 
                          aggfunc='sum').plot.bar(subplots = True,
                                                  layout=(2,2),
                                                   figsize = (10,10),
                                                   sharey = True))
```

```python
(transactions.groupby(['date']).agg({'SALES_VALUE':'sum'}).loc['2016-04':'2017-10']
             .resample('M').sum().plot())
```

```python
df1  = (transactions.groupby(['date']).agg({'SALES_VALUE':'sum'}).resample('M').sum())
df1
```

```python
df1.assign(
    difference = df1.diff(12)).plot()
```

```python
df1.assign(
    difference = df1['SALES_VALUE'] - df1['SALES_VALUE'].shift(12))
```

```python
df1
```

```python
df1.assign(
    growth_percent = (df1['SALES_VALUE'] /df1['SALES_VALUE'].shift(1)).sub(12).mul(100).round()
    ).plot()
```

```python
(transactions.pivot_table(index= transactions['date'].dt.dayofweek,
                          columns = transactions['date'].dt.year,
                          values = 'SALES_VALUE',
                          aggfunc= 'sum')).plot()
                    
```

```python
(transactions.pivot_table(index= transactions['date'].dt.dayofweek,
                          columns = transactions['date'].dt.year,
                          values = 'SALES_VALUE',
                          aggfunc= 'sum').T).apply(lambda x  : x / sum(x) *100, axis = 1).plot.bar(stacked = True)
```

```python
(transactions.pivot_table(index= transactions['date'].dt.dayofweek,
                          columns = transactions['date'].dt.year,
                          values = 'SALES_VALUE',
                          aggfunc= 'sum')).apply(lambda x  : x / sum(x) *100, axis = 1).plot.bar(stacked = True)
```

```python
demographics = pd.read_csv('hh_demographic.csv', 
                           usecols= ['AGE_DESC', 'INCOME_DESC', 'household_key','HH_COMP_DESC'],
                           dtype= {'household_key': 'int16', 'AGE_DESC' :'category', 'HH_COMP_DESC':'category', 'INCOME_DESC':'category'})
demographics
```

```python
demographics.describe(include= 'all')
```

```python
demographics.info()
```

```python
transactions.merge(demographics,
                   on = ['household_key'],
                   how= 'left').groupby(['household_key'])[['SALES_VALUE']].sum()
```

```python
transactions = transactions.merge(demographics,
                   on=['household_key'])
transactions.head()
```

```python
transactions.groupby(['AGE_DESC_x'])['SALES_VALUE'].sum().sort_values().plot.bar()
```

```python
transactions.groupby(['AGE_DESC_x'])['SALES_VALUE'].sum().sort_values().plot()
```
